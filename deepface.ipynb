```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
foto="hadise.jpg"
img=cv2.imread(foto)

plt.imshow(img[:,:, ::-1])
plt.show

resp=DeepFace.analyze(foto)

print(resp['age'])
print(resp['gender'])
print(resp['dominant_emotion'])
print(resp['dominant_race'])
```
